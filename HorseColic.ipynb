```python
# pandas dataframes
df = pd.concat([x, y], axis=1)
df.head()
df.describe(include='all')
```

```python
# checking the missing values
df.info()
df.isnull().sum()
```

```python
#handling missing values with mean
df = df.fillna(df.mean(numeric_only=True))
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
outcome_mapping = {
    1.0: 'Recovered',
    2.0: 'Euthanized',
    3.0: 'Died'
}
```

```python
df['outcome_label'] = df['outcome'].map(outcome_mapping)
df[['outcome', 'outcome_label']].head()
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x='outcome_label', data=df,
                   order=['Recovered', 'Euthanized', 'Died'])

plt.title('Distribution of Outcome Classes')
plt.xlabel('Outcome Class')
plt.ylabel('Count')
```
