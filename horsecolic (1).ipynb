```python
from ucimlrepo import fetch_ucirepo, list_available_datasets
```

```python
#import the dataset of horse holic
horse_colic = fetch_ucirepo(id=47)
```

```python
#access data
x = horse_colic.data.features  #input features
y = horse_colic.data.targets   #output (outcome for each )
print(x.head())
print(y.head())
```

```python
%matplotlib inline
```

```python
print(x.head())
print(y.head())
print(x.info())
```

```python
# pandas dataframes
df = pd.concat([x, y], axis=1)
df.head()
df.describe(include='all')
```

```python
# checking the missing values
df.info()
df.isnull().sum()
```

```python
#handling missing values with mean
df = df.fillna(df.mean(numeric_only=True))
df.isnull().sum()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```

```python
print(df.isna().sum())

print(df.info())
print(df.describe(include='all'))
print(df.head())
```

```python
outcome_mapping = {
    1.0: 'Recovered',
    2.0: 'Euthanized',
    3.0: 'Died'
}
```

```python
df['outcome_label'] = df['outcome'].map(outcome_mapping)
df[['outcome', 'outcome_label']].head()
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x='outcome_label', data=df,
                   order=['Recovered', 'Euthanized', 'Died'])

plt.title('Distribution of Outcome Classes')
plt.xlabel('Outcome Class')
plt.ylabel('Count')
```

```python
df['surgical_lesion_label'] = df['surgical_lesion'].map({1.0: 'No Surgery (1)', 2.0: 'Surgery (2)'})
```

```python
# Plot distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='surgical_lesion_label', data=df,)
plt.title('Distribution of Surgical Lesion (Target Variable)')
plt.xlabel('Surgical Lesion Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
